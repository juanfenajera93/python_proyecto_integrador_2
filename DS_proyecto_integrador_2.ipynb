## Información general del estudio

A continuación, se presenta información a tomar en cuenta sobre el proceso en general. 

El proceso de extracción de oro del mineral se realiza en varias etapas. Inicialmente, el mineral extraído se somete a un tratamiento primario para obtener la mezcla de mineral, conocida como alimentación *rougher*, que es la materia prima para la etapa de flotación.

- **Flotación:** En esta etapa, la mezcla de mineral de oro se introduce en las plantas de flotación para separar y obtener un concentrado de oro (llamado concentrado *rougher*) y los residuos del proceso (llamados colas *rougher*). Este proceso puede verse afectado por factores como la volatilidad y el estado físico-químico de la pulpa de flotación.

- **Purificación:** Posteriormente, el concentrado *rougher* se somete a dos etapas de purificación para eliminar impurezas restantes y obtener el concentrado final de oro. Este proceso también genera nuevos residuos, conocidos como colas.

Este procedimiento detallado es fundamental para nuestro análisis, ya que nos proporciona una visión clara de cómo se extrae el oro del mineral y cómo se pueden optimizar estos procesos.

### Descripción de datos

Aquí se presenta una descripción general de los datos que se obtendrán en los archivos. 

**Proceso tecnológico**
- `Rougher feed` — materia prima
- `Rougher additions` (o adiciones de reactivos) - reactivos de flotación: xantato, sulfato, depresante
    - `Xantato` — promotor o activador de la flotación
    - `Sulfato` — sulfuro de sodio para este proceso en particular
    - `Depresante` — silicato de sodio
- `Rougher process` — flotación
- `Rougher tails` — residuos del producto
- `Float banks` — instalación de flotación
- `Cleaner process` — purificación
- `Rougher Au` — concentrado de oro rougher
- `Final Au` — concentrado de oro final

**Parámetros de las etapas**
- `air amount` — volumen de aire
- `fluid levels` 
- `feed size` — tamaño de las partículas de la alimentación
- `feed rate`

**Denominación de las características**

Así es como se denominan las características:
`[stage].[parameter_type].[parameter_name]`

Ejemplo: `rougher.input.feed_ag`

Valores posibles para `[stage]`:
- `rougher` — flotación
- `primary_cleaner` — purificación primaria
- `secondary_cleaner` — purificación secundaria
- `final` — características finales

Valores posibles para `[parameter_type]`:
- `input` — parámetros de la materia prima
- `output` — parámetros del producto
- `state` — parámetros que caracterizan el estado actual de la etapa
- `calculation` — características de cálculo

### Cálculo de recuperación

Para verificar el cálculo de la recuperación, necesitaremos la fórmula para el cálculo de la recuperación, que es la siguiente: 

`recovery = C*(F-T) / F*(C-T) * 100%`

donde:

- `C` es la concentración del oro en el concentrado después de la flotación (`rougher.output.concentrate_au`)
- `F` es la concentración del oro en el feed/entrada al proceso antes de la flotación (`rougher.input.feed_au`)
- `T` es la concentración del oro en la salida del proceso, después de la flotación (`rougher.output.tail_au`)

En base a esta fórmula, se define una función para calcular la recuperación para llamarlo `calculate_recovery` el cual tiene como argumenteos `C`, `F` y `T`. 

```python
# función para calcular la recuperación
def calculate_recovery(C, F, T):
    recovery = ((C * (F - T)) / (F * (C - T))) * 100
    return recovery
```

Se aplicará esta función para las columnas `rougher.output.concentrate_au` que corresponde a `C`, `rougher.input.feed_au` que corresponde a `F` y `rougher.output.tail_au` que corresponde a `T`. 

```python
# aplicación de función
calculated_recovery = calculate_recovery(train['rougher.output.concentrate_au'], 
                                         train['rougher.input.feed_au'], 
                                         train['rougher.output.tail_au'])
```

A partir de esto, se filtrará la información donde no existan `NaN`, esto para obtener solo valores válidos: 

```python
# filtro sin NaN
filter_wo_nan = ~np.isnan(calculated_recovery) & ~np.isnan(train['rougher.output.recovery'])
```

Ahora procedemos a calcular el error absoluto medio `MAE` entro los valores de recuperación calculados y los reales. Para esto, aplicamos el filtro anterior para no tomar en cuenta los `NaN`. 

```python
# cálculo de MAE
mae = mean_absolute_error(train.loc[filter_wo_nan, 'rougher.output.recovery'], calculated_recovery[filter_wo_nan])
# resultado
print(f'Error Absoluto Medio (MAE) entre los valores de recuperación calculados y los valores de recuperación reales: {mae}')
```

Tenemos un valor de MAE bastante pequeño, esto significa que las predicciones del modelo (en este caso, los valores de recuperación calculados) están muy cerca de los valores reales. Es decir, la fórmula que se utilizó para calcular la recuperación está dando resultados muy precisos en comparación con valores de recuperación reales. 

Y esto, para el contexto de minería de oro, nos ayuda a simular el proceso de recuperación del oro en la extracción de un mineral, lo cual es un buen resultado para poder predecir los datos.

## Construcción del modelo

### Función `sMAPE`

El `sMAPE` es el error medio absoluto porcentual simétrico, el cual es similar al MAE, pero se expresa en valores relativos en lugar de absolutos. En base a la información provista, se creará una función que ayudará a calcular el `sMAPE`. Esto nos ayudará al momento de aplicar los modelos de aprendizaje supervisado. 

```python
# función `sMAPE`
def sMAPE(y_true, y_pred):
    numerator = np.abs(y_true - y_pred)
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    return np.mean(numerator / denominator) * 100
```

A partir de la función de `sMAPE` como una métrica de scoring personalizada, se puede usar en la función `cross_val_score`. Para esto, se puede usar con la función `make_scorer` de `sklearn.metrics`. 

```python
# scorer personalizado para la función cross_val_score
smape_score = make_scorer(sMAPE, greater_is_better=False)
```

### Entrenamiento de modelos

Vamos a realizar el entrenamiento de modelos. Para esto, se utilizará el dataset `new_train`. Vamos a limpiar la columna `date` que no se utilizará en el análisis. 
- `date`

Ahora se definirá las siguientes variables como variables objetivo: 
- `rougher.output.recovery`
- `final.output.recovery`

```python
# definición de X_train y y_train
features = new_train.columns.drop(['date', 'rougher.output.total', 'final.output.total'])
X_train = new_train[features]
y_train_rougher = train['rougher.output.recovery']
y_train_final = train['final.output.recovery']
```

Los valores de las variables objetivos se habían eliminado anteriormente, por lo que se utilizarán a partir de `train`. Las variables tienen valores ausentes, por lo que se rellenarán a partir de la media.

```python
# reemplazo valores ausentes
y_train_rougher = y_train_rougher.fillna(y_train_rougher.mean())
y_train_final = y_train_final.fillna(y_train_final.mean())
```

Ahora creamos los modelos, vamos a utilizar el modelo de regresión lineal y el árbol de decisión. 

```python
# implementación de modelos
linear_model_rougher = LinearRegression()
linear_model_final = LinearRegression()
tree_model_rougher = DecisionTreeRegressor()
tree_model_final = DecisionTreeRegressor()

# entrenamiento de modelos
linear_model_rougher.fit(X_train, y_train_rougher)
linear_model_final.fit(X_train, y_train_final)

tree_model_rougher.fit(X_train, y_train_rougher)
tree_model_final.fit(X_train, y_train_final)
```

Procedemos a entrenar los modelos: 

```python
# evaluación de modelos con cross_val_score
linear_scores_rougher = -cross_val_score(linear_model_rougher, X_train, y_train_rougher, cv=5, scoring=smape_score)
linear_scores_final = -cross_val_score(linear_model_final, X_train, y_train_final, cv=5, scoring=smape_score)

tree_scores_rougher = -cross_val_score(tree_model_rougher, X_train, y_train_rougher, cv=5, scoring=smape_score)
tree_scores_final = -cross_val_score(tree_model_final, X_train, y_train_final, cv=5, scoring=smape_score)

print(f"sMAPE para el modelo lineal (rougher): {np.mean(linear_scores_rougher)}")
print(f"sMAPE para el modelo lineal (final): {np.mean(linear_scores_final)}")

print(f"sMAPE para el modelo de árbol de decisión (rougher): {np.mean(tree_scores_rougher)}")
print(f"sMAPE para el modelo de árbol de decisión (final): {np.mean(tree_scores_final)}")
```

En base a estos valores, podemos medir el rendimiento del modelo con las puntuaciones del `sMAPE` el cual muestra el error medio porcentual de los modelos. En este caso, cuanto más baja sea la puntuación, mejor será el rendimiento del modelo. 

De acuerdo con los resultados, los modelos de árbol de decisión parecen superar a los modelos de regresión lineal para ambas etapas del proceso (rougher y final). Específicamente, el modelo de árbol de decisión tiene un sMAPE de 0.555 para la etapa rougher y 0.201 para la etapa final, mientras que los modelos de regresión lineal tienen un sMAPE más alto de 0.874 y 0.641 respectivamente. Por lo tanto, parece que los modelos de árbol de decisión pueden ser más adecuados para este problema en particular.

Además, para ambos tipos de modelos, la etapa rougher del proceso tiene un error `sMAPE` más alto que la etapa final. Esto puede indicar que la etapa rougher del proceso es más difícil de predecir con precisión, lo cual podría deberse a una mayor complejidad o variabilidad en los datos correspondientes a esta etapa.

Por lo tanto, según los resultados de sMAPE, el modelo de árbol de decisión tiene un rendimiento superior al modelo lineal tanto en los casos `rougher` como `final`, ya que tiene valores de sMAPE más bajos en ambos casos.
